# EJERCICIO 1
Escriba un programa que determine si un estudiante puede
acceder a una beca académica, en función de su promedio y su
nivel de estudios.

```python
# Variable global opcional
pais = "Ecuador"

def beca_estudiante():
    # Solicitar datos al usuario
    promedio = float(input("Ingrese el promedio académico: "))  #Captura el promedio académico como número real.
    nivel = input("Ingrese el nivel de estudios (Basico, Bachillerato, Universitario): ")  #Captura el nivel de estudios.

    print("**********DATOS DEL ESTUDIANTE***********")
    print("Promedio: ", promedio)
    print("Nivel de estudios: ", nivel)
    print("Pais: ", pais)

    # Uso de operadores ternarios para evaluar condiciones

    resultado = (                                                               #Uso de operadores ternarios para evaluar las condiciones de la beca.
        "Obtiene la beca" if promedio >= 9.0 and nivel == "Universitario" else
        "Obtiene la beca" if 8.0 <= promedio <= 8.9 and nivel == "Bachillerato" else
        "No obtiene la beca"
    )

    # Mostrar resultado
    print("Resultado: ", resultado)    #Muestra si el estudiante obtiene o no la beca.
    print("*****************************************")

# Programa principal
beca_estudiante()
```

# EJERCICIO 2

 Escriba un programa que calcule el valor final de una matrícula,
basado en el costo base y el tipo de estudiante, utilizando
operadores ternarios.

```python
# variable

def calcular_matricula():
    # Solicitar datos al usuario
    costo_base = float(input("Ingrese el costo base de la matrícula: "))
    tipo = input("Ingrese el tipo de estudiante (Regular, Becado, Extraordinario): ")

    print("**********DATOS DE MATRÍCULA***********")
    print("Costo base: ", costo_base)
    print("Tipo de estudiante: ", tipo)

    # Aplicar descuento si corresponde (10% si costo base > 1500)

    descuento = costo_base * 0.10 if costo_base > 1500 else 0        #Aplica el descuento del 10% si el costo base supera 1500.
    costo_con_descuento = costo_base - descuento

    # Determinar recargo según tipo de estudiante usando operador ternario


    recargo = (                                                      # Usa operadores ternarios para asignar el porcentaje de recargo según el tipo de estudiante.
        costo_con_descuento * 0.15 if tipo == "Regular" else
        costo_con_descuento * 0.05 if tipo == "Becado" else
        costo_con_descuento * 0.25 if tipo == "Extraordinario" else 0
    )

    # Calcular valor final
    valor_final = costo_con_descuento + recargo                      #Calcula el valor final de la matrícula.

    # Mostrar resultados
    print("Descuento aplicado: ", descuento)
    print("Recargo aplicado: ", recargo)
    print("Valor final a pagar: ", valor_final)
    print("*****************************************")

# Programa principal
calcular_matricula()
```

# EJERCICIO 3
Escriba un programa en C++ que evalúe la velocidad de un
vehículo, con base en una velocidad ingresada en km/h.

```python
# variable
velocidad_kmh = 0
velocidad_ms = 0
clasificacion = ""

def evaluar_velocidad():
    # Solicitar velocidad al usuario
    velocidad_kmh = float(input("Ingrese la velocidad en km/h: "))   #Captura la velocidad ingresada como número decimal.

    # Verificar si la velocidad es válida

    if velocidad_kmh < 0 or velocidad_kmh > 300:      #Verifica que la velocidad esté en el rango válido
        print("Error: La velocidad ingresada no es válida (debe estar entre 0 y 300 km/h).")
    else:

        # Convertir a m/s
        velocidad_ms = velocidad_kmh / 3.6    #Convierte la velocidad de km/h a m/s.

        # Clasificación usando operadores ternarios anidados

        clasificacion = (                                       #Usa operadores ternarios anidados para clasificar la velocidad.
            "Lenta" if velocidad_kmh < 40 else
            "Moderada" if 40 <= velocidad_kmh <= 80 else
            "Rápida" if 80 < velocidad_kmh <= 120 else
            "Muy rápida"
        )

        # Mostrar resultados
        print("Velocidad ingresada: ", velocidad_kmh, "km/h")
        print("Velocidad convertida: ", round(velocidad_ms, 2), "m/s")
        print("Clasificación: ", clasificacion)

# Programa principal
evaluar_velocidad()
```

# EJERCICIO 4
Escriba un programa que defina un número secreto fijo (porejemplo, 9), o lo puede ingresar por teclado.
El usuario tiene hasta 5 intentos para adivinar el número.

Por cada intento:

    Se incrementa un contador de intentos.
    Se compara el número ingresado con el secreto:
    Si es igual: se imprime “¡Correcto! Lo lograste en X intentos” y el programa termina.

```python
def adivinar_numero():
    # Número secreto fijo (puede cambiarse o pedirse por teclado)

    numero_secreto = 9   # Define el número secreto fijo.
    intentos = 0
    max_intentos = 5

    print("Tienes hasta 5 intentos para adivinar el número secreto.")

    while intentos < max_intentos:    # Controla el ciclo de intentos (máximo 5).
        numero = int(input("Ingrese un número: "))
        intentos += 1

        if numero == numero_secreto:     # Verifica si el número ingresado coincide con el secreto.

            print(f"¡Correcto! Lo lograste en {intentos} intentos")
            return  # Termina el programa si acierta
        else:
            # Operador ternario para indicar si es mayor o menor

            mensaje = "Demasiado alto" if numero > numero_secreto else "Demasiado bajo" # Usa operador ternario para dar la pista.
            print(mensaje)

    # Si se agotan los intentos
    print(f"Lo siento, se agotaron los intentos. El número era {numero_secreto}")    # Mensaje final si se agotan los intentos.

# Programa principal
adivinar_numero()
```

# Preguntas:  
**1. ¿Por qué es importante elegir el tipo de dato adecuado al declarar
una variable en C++?**

    Porque el tipo de dato define cómo se guarda la información en memoria y qué operaciones se pueden hacer con ella. Si eliges mal, puedes desperdiciar recursos o tener errores inesperados. Por ejemplo, usar int para almacenar decimales no tiene sentido, y usar double para contar personas tampoco. Es como escoger el recipiente correcto para lo que quieres guardar.

**2. ¿En qué situaciones es útil el operador ternario (? :) y en cuáles sería
mejor usar if-else?**

    El ternario es muy práctico cuando quieres tomar una decisión rápida y asignar un valor en una sola línea, por ejemplo:
    mensaje = (edad >= 18) ? "Mayor de edad" : "Menor de edad";
    Pero si la lógica es más compleja, con varias condiciones o bloques de código, el if-else es más claro y fácil de leer. En resumen: ternario para lo simple, if-else para lo elaborado.

**3. ¿Qué diferencia hay entre una variable local y una global, y por
qué se recomienda evitar las globales?**

    La local existe solo dentro de la función donde se declara, mientras que la global está disponible en todo el programa. El problema de las globales es que cualquiera puede modificarlas, lo que puede generar errores difíciles de rastrear. Por eso se recomienda usarlas poco y preferir las locales, que son más seguras y ordenadas.

**4. Al operar con tipos distintos (ej. int y double), ¿qué ocurre en C++?
¿Es siempre seguro?**

    C++ hace una conversión automática (lo que se llama promoción de tipos). Por ejemplo, si divides un int entre un double, el resultado será double. Generalmente funciona bien, pero no siempre es seguro: puedes perder precisión o tener resultados inesperados si no controlas el tipo final. Por eso conviene ser explícito y usar casting cuando lo necesitas.

**5. ¿Qué error ocurre si usas = en lugar de == en una comparación?
¿Cómo lo detectarías?**

    El = asigna un valor, mientras que == compara. Si escribes if (x = 5) en lugar de if (x == 5), lo que haces es asignar 5 a x, y la condición siempre será verdadera (salvo que asignes 0). Es un error muy común. Se detecta porque el programa se comporta raro, y muchos compiladores modernos te avisan con una advertencia. Una buena práctica es escribir la constante primero: if (5 == x), así si olvidas un =, el compilador te marca error.
